```python
import pygame
from tkinter import *
from tkinter import filedialog
win = Tk()
import os

win.maxsize(width= 300, height= 300)
win.minsize(width= 300, height= 300)
win.title("MP3 Music Player")

pygame.mixer.init()
running = ""

def show_list():
    #listbox.delete(0,END)
    for i in songs:
        listbox.insert(END,i)

def load_song():
    file_path= filedialog.askdirectory()
    os.chdir(file_path)
    global songs
    songs = os.listdir()
   # print(songs)
    show_list();
    current_song.config(text="Choose a song")

def volume(v):
    alert.config(text="")
    pygame.mixer.music.set_volume(volume_slider.get())
    
def play_music():
    alert.config(text="")
    global running
    running = listbox.get(ACTIVE)
    pygame.mixer.music.load(running)
    pygame.mixer.music.play()
    current_song.config(text = running)
    listbox.delete(0,END)
    show_list()
    listbox.select_set(0)
    play_button.config(text="Restart")

    
def pause_music():
    alert.config(text="")
        
    if(pygame.mixer.music.get_busy()):
        pygame.mixer.music.pause()
        pause_button.config(text="Continue")
    else:
        pygame.mixer.music.unpause()
        pause_button.config(text="Pause")
    
def stop_music():
    alert.config(text="")
    pygame.mixer.music.stop()
    pause_button.config(text="Pause")
    current_song.config(text="Choose a Song")
    listbox.delete(0,END)
    show_list()
    listbox.select_set(0)
    play_button.config(text="Play")
    
def next_music():
    alert.config(text="")
    global running
    playing_song = running
    index_song = songs.index(playing_song)
   # print(index_song,running)
   # print(len(songs))
    if index_song < len(songs)-1:
        index_song = index_song + 1
        playing_song = songs[index_song]
        #print(playing_song)
        pygame.mixer.music.load(playing_song)
        pygame.mixer.music.play()

        listbox.delete(0,END)
        show_list();
        listbox.select_set(index_song)
        running = playing_song
        current_song.config(text = playing_song)
    else:
        listbox.delete(0, END)
        show_list()
        listbox.select_set(0)
        play_music();
        
def previous_song():
    global running
    playing_song = running
    index_song = songs.index(playing_song)
    #print(index_song)
    if index_song != 0:
        index_song = index_song - 1
        
        playing_song = songs[index_song]
        pygame.mixer.music.load(playing_song)
        pygame.mixer.music.play()
        running = playing_song
        listbox.delete(0,END)
        show_list();
        listbox.select_set(index_song)
    else:
        alert.config(text="No song is there")
        
        
    
button = Button(win, width=10, text="load songs", bg="red", command=load_song)
button.place(x=1,y=0)

current_song = Label(win, text = "Load songs")
current_song.place(x = 95, y = 0)

listbox= Listbox(win, selectmode= SINGLE, width = 49, fg="white", bg="blue")
listbox.place(x=0,y=23)

scroll = Scrollbar(win,bg="black")
scroll.place(x=280,y=23)
listbox.config(yscrollcommand=scroll.set)
scroll.config(command= listbox.yview)

previous_button = Button(win, width=10,text = "Previous\nSong", bg ="red", command= previous_song)
previous_button.place(x=2,y=200)

play_button = Button(win, width=10,text = "Play", bg ="red", command= play_music)
play_button.place(x=100,y=190)

pause_button = Button(win, width=10,text = "Pause", bg ="red", command= pause_music)
pause_button.place(x=100,y=220)


next_button = Button(win, width=10,text = "Next\nSong", bg ="red", command = next_music )
next_button.place(x=200,y=200)


stop_button = Button(win, width=10,text = "Stop", bg ="red", command= stop_music )
stop_button.place(x=210,y=260)

#volume_up_button = Button(win, width=10,text = "Volume Up", bg ="red" )
#volume_up_button.place(x=2,y=250)

#volume_down_button = Button(win, width=12,text = "Volume Down", bg ="red" )
#volume_down_button.place(x=100,y=250)
alert = Label(win, text = "")
alert.place(x = 2, y = 245)

volume_slider = Scale(win, orient=HORIZONTAL, from_=0, to_=100 ,command=volume)
volume_slider.set(80)
volume_slider.place(x=100,y=250)
#print(volume_slider.get())
lable = Label(text="VOLUME:-")
lable.place(x=44,y=269)

def on_closing():
    pygame.mixer.music.stop()
    pygame.mixer.music.unload()
    win.destroy()

win.protocol("WM_DELETE_WINDOW", on_closing)
win.mainloop()
```
